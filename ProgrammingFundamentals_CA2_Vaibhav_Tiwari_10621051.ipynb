Data cleaning: Replaced the empty (-) fields with 0 in the dataset.

```python
for column in data.columns :
    entire_column_dataload_arr = []
    for row_data in data[column] :

        
        if str(row_data).__contains__("-") :
            format_row_data = int(0)
            entire_column_dataload_arr.append(format_row_data)
        else :
            entire_column_dataload_arr.append(row_data)
    data[column] = entire_column_dataload_arr

data
```

```python
# Find range of matches played by all players in T20 format in the year 2022
minumum_matches_played = 0
maximum_matches_played = 0
loop_count = 0

for value in data["Matches"] :
  
  row_value = str(value).replace(",","")
  
  if loop_count == 0 :
    minumum_matches_played = float(row_value)

  if float(minumum_matches_played) > float(row_value) :
    minumum_matches_played = float(row_value)
  elif float(maximum_matches_played) < float(row_value) :
    maximum_matches_played = float(row_value)

  loop_count = loop_count + 1

minumum_matches_played = str(minumum_matches_played)
maximum_matches_played = str(maximum_matches_played)

print("Range of matches played is in between",minumum_matches_played, "and", maximum_matches_played)
```

```python
# Find the batsman with maximum strike rate in T20 2022

maximum_sr = 0.0
index = 0
count = 0
for sr in data["Batting Strike Rate"]:
  if count == 0 :
    maximum_sr = float(sr)
    index = count
  else :
    if maximum_sr <= float(sr) :
      maximum_sr = float(sr)
      index = count
  count = count+1

print(data["Player Name"][index], "has the maximum SR",maximum_sr)
```

```python
# Find the batsman with maximum batting average in T20 2022

maximum_avg = 0.0
index = 0
count = 0
for x in data["Batting Average"]:
  if count == 0 :
    maximum_avg = float(x)
    index = count
  else :
    if maximum_avg <= float(x) :
      maximum_avg = float(x)
      index = count
  count = count+1

print(data["Player Name"][index], "has the maximum batting average",maximum_avg)
```
